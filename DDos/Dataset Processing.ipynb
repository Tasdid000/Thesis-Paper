```python
label_flowid = LabelEncoder()

label_flowid.fit(data_X['Flow ID'])
data_X['Flow ID'] = label_flowid.fit_transform(data_X['Flow ID'])

label_SourceIP = LabelEncoder()

label_SourceIP.fit(data_X[' Source IP'])
data_X[' Source IP'] = label_SourceIP.fit_transform(data_X[' Source IP'])

label_DestinationIP = LabelEncoder()

label_DestinationIP.fit(data_X[' Destination IP'])
data_X[' Destination IP'] = label_DestinationIP.fit_transform(data_X[' Destination IP'])

le_timestamp = LabelEncoder()
le_timestamp.fit(data_X[' Timestamp'])
data_X[' Timestamp'] = le_timestamp.fit_transform(data_X[' Timestamp'])

data_X
```
